```python
import cv2                             #All librarys
import numpy as np
from matplotlib import pyplot as plt
import math
from time import sleep
import datetime
from time import time
import time
from datetime import date
```

```python

img =np.zeros((1000,1000,3),np.uint8)
cv2.namedWindow('image')                         #Name



while 1:
  date_time_now = datetime.datetime.now()
  time_now = date_time_now.time()
    
  hour = math.fmod(time_now.hour, 12)
  minute = time_now.minute
  second = time_now.second

  second_iti = math.fmod(second * 6 + 270, 360)
  minute_iti = math.fmod(minute * 6 + 270, 360)
  hour_iti = math.fmod((hour * 30) + (minute / 2) + 270, 360)
  img =np.zeros((1000,1000,3),np.uint8)           #background
  
  cv2.circle(img,(650,400), 250, (255,255,0),10)  #circlr
  
  x=round(650+190*math.cos(second_iti*3.14/180))    #with current time
  y=int(400+190*math.sin(second_iti*3.14/180))
  cv2.line(img,(650,400),(x,y),(0,0,255),1)
  
  x=int(650+190*math.cos(minute_iti*3.14/180))
  y=int(400+190*math.sin(minute_iti*3.14/180))
  cv2.line(img,(650,400),(x,y),(255,0,255),3)
  
  x=int(650+180*math.cos(hour_iti*3.14/180))
  y=int(400+180*math.sin(hour_iti*3.14/180))
  cv2.line(img,(650,400),(x,y),(0,255,0),5)
  
 
  
  x=int(650+250*math.cos(math.radians(0)))  #points in clock
  y=int(400+250*math.sin(math.radians(0)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  x=int(650+250*math.cos(math.radians(30)))
  y=int(400+250*math.sin(math.radians(30)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  x=int(650+250*math.cos(math.radians(60)))
  y=int(400+250*math.sin(math.radians(60)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  x=int(650+250*math.cos(math.radians(90)))
  y=int(400+250*math.sin(math.radians(90)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  x=int(650+250*math.cos(math.radians(120)))
  y=int(400+250*math.sin(math.radians(120)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  x=int(650+250*math.cos(math.radians(150)))
  y=int(400+250*math.sin(math.radians(150)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  x=int(650+250*math.cos(math.radians(270)))
  y=int(400+250*math.sin(math.radians(270)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  x=int(650+250*math.cos(math.radians(300)))
  y=int(400+250*math.sin(math.radians(300)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  x=int(650+250*math.cos(math.radians(330)))
  y=int(400+250*math.sin(math.radians(330)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  x=int(650+250*math.cos(math.radians(180)))
  y=int(400+250*math.sin(math.radians(180)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  x=int(650+250*math.cos(math.radians(210)))
  y=int(400+250*math.sin(math.radians(210)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  x=int(650+250*math.cos(math.radians(240)))
  y=int(400+250*math.sin(math.radians(240)))
  cv2.circle(img,(x,y), 10, (255,0,0),-1)
  
  font = cv2.FONT_HERSHEY_SIMPLEX             #tipe of font
  datet = str(datetime.datetime.now())        #link with the current time
  cv2.putText(img, 'date:          time:', (50,70), font, 1, (225,225,0), 2 )  #Text change with time
  cv2.putText(img,  datet, (50,120), font, 1, (225,225,0), 2 )


  cv2.imshow('image',img)             #show all in window
  k = cv2.waitKey(1)                  # wait time
  if k == ord('q'):
        break 
cv2.destroyAllWindows()               #close when finish
```
